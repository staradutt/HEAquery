```python
import pickle
with open(chunks_path, "rb") as f:
    chunks = pickle.load(f)

print(f"Loaded {len(chunks)} chunks")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
import torch
from langchain.docstore.document import Document

# Optional: attach metadata like filename or section if you have it
documents = [Document(page_content=c) for c in chunks]
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-base-en",
    model_kwargs={"device": device}  # This makes it use GPU
)

# Step 4: Embed in batches with progress bar and save intermediate results
from langchain.vectorstores import FAISS
from tqdm import tqdm


import math

BATCH_SIZE = 1000  # adjust depending on RAM / GPU
INDEX_SAVE_PATH = '/content/drive/MyDrive/faiss_index_hugface_BAAI_new'
faiss_index = None
num_batches = math.ceil(len(documents) / BATCH_SIZE)

for i in range(num_batches):
    batch_docs = documents[i*BATCH_SIZE : (i+1)*BATCH_SIZE]
    print(f"Processing batch {i+1}/{num_batches} with {len(batch_docs)} docs...")

    if faiss_index is None:
        faiss_index = FAISS.from_documents(batch_docs, embeddings)
    else:
        faiss_index.add_documents(batch_docs)  # <-- embeddings not needed here

    # Save intermediate progress every batch
    faiss_index.save_local(INDEX_SAVE_PATH)
    print(f"Saved FAISS index at batch {i+1}")

print("FAISS index creation complete!")
```
